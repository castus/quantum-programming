```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
import runner

q = QuantumRegister(3, name="quantum")
c = ClassicalRegister(3, name="classical")
circuit = QuantumCircuit(q, c, name="circuit")
circuit.id(q[0])
circuit.id(q[1])
circuit.id(q[2])
circuit.x(q[0]) # Uncomment this to have 4 values
# circuit.h(q[0]) # Uncomment this to have 8 values
circuit.h(q[1])
circuit.h(q[2])
circuit.measure(q, c)

runner.run(circuit)
circuit.draw("mpl")
```
