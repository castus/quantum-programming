```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
import runner

q1 = QuantumRegister(1, name="quantum 1")
q2 = QuantumRegister(1, name="quantum 2")
q3 = QuantumRegister(1, name="quantum 3")
c1 = ClassicalRegister(1, name="classical 1")
qc1 = QuantumCircuit(q1, q2, q3, c1, name="circuit 1")

qc1.id(0)
qc1.id(1)
qc1.id(2)
qc1.h(2)
qc1.cswap(2, 0, 1)
qc1.h(2)
qc1.x(2)

qc1.measure(q3, c1)
task = runner.run(qc1, shots=1000)
qc1.draw("mpl")
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(task.result().get_counts())
```
