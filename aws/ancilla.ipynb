```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
import runner

q = QuantumRegister(7, name="quantum")
c = ClassicalRegister(7, name="classical")
circuit = QuantumCircuit(q, c, name="circuit")

circuit.h(q[0])
circuit.h(q[1])
circuit.h(q[2])
circuit.h(q[3])
circuit.id(q[4])
circuit.id(q[5])
circuit.id(q[6])

circuit.ccx(q[0], q[1], q[4])
circuit.ccx(q[2], q[3], q[5])
circuit.ccx(q[4], q[5], q[6])
circuit.cz(q[6], q[0])

circuit.measure(q, c)

runner.run(circuit)
circuit.draw("mpl")
```
