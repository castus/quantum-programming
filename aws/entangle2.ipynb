```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
import runner

q = QuantumRegister(3, name="quantum")
c = ClassicalRegister(3, name="classical")
qc = QuantumCircuit(q, c, name="circuit")
qc.h(q[0])
qc.h(q[1])
qc.h(q[2])
qc.measure(q, c)

runner.run(qc)
qc.draw("mpl")
```
